```python
from Crypto.Cipher import AES
from Crypto.Util.Padding import pad, unpad
from Crypto.Random import get_random_bytes

# Generate a random secret key (must be 16, 24, or 32 bytes long for AES)
key = get_random_bytes(16)

# Generate a random Initialization Vector (IV) (16 bytes for AES)
iv = get_random_bytes(16)

# Data to encrypt (plaintext)
plaintext = b"Hello, this is a secret message!"

# Create AES cipher in CBC mode
cipher = AES.new(key, AES.MODE_CBC, iv)

# Pad the plaintext to make its length a multiple of the block size (16 bytes for AES)
padded_plaintext = pad(plaintext, AES.block_size)

# Encrypt the padded plaintext
ciphertext = cipher.encrypt(padded_plaintext)

print(f"Key: {key.hex()}")
print(f"IV: {iv.hex()}")
print(f"Ciphertext: {ciphertext.hex()}")
```

```python
# Create AES cipher in CBC mode for decryption
decipher = AES.new(key, AES.MODE_CBC, iv)

# Decrypt the ciphertext
decrypted_padded_plaintext = decipher.decrypt(ciphertext)

# Unpad the decrypted plaintext
decrypted_plaintext = unpad(decrypted_padded_plaintext, AES.block_size)

print(f"Decrypted Plaintext: {decrypted_plaintext.decode()}")
```

```python
from cryptography.fernet import Fernet
from PIL import Image
import matplotlib.pyplot as  plt

#generate a key
key = Fernet.generate_key()
obj = Fernet(key)

image = Image.open('/content/img.jpeg')
image_data = image.tobytes() # convert to bytes because Fernet uses bytes format
plt.imshow(image)
```

```python
encrypted_data = obj.encrypt(image_data) #data must be in bytes

with open('encrypted.jpeg','wb') as enc:
    enc.write(encrypted_data)

#Our image is encrypted now and data is in bytes. In order to display it, make image from bytes
encrypted_image = Image.frombytes(image.mode,image.size,encrypted_data)
plt.imshow(encrypted_image)
```

```python
decrypted_data = obj.decrypt(encrypted_data)
decrypted_image = Image.frombytes(image.mode,image.size,decrypted_data)
plt.imshow(decrypted_image)
```
